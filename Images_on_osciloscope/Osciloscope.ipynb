```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys
np.set_printoptions(threshold=sys.maxsize)

#Parameters
DAC_MAX = 4095
image = cv2.imread("kotek_dark_maly.png", cv2.IMREAD_GRAYSCALE)
threshold = 100

_, binary_image = cv2.threshold(image, threshold, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, 'gray')
plt.show()

y, x = np.where(binary_image == 255)
y = (DAC_MAX - (y / np.shape(binary_image[0]) * DAC_MAX)).astype(np.int16)
x = (x / np.shape(binary_image[1]) * DAC_MAX).astype(np.int16)

pairs = list(zip(x, y))
sorted_pairs = sorted(pairs, key=lambda pair: pair[0])
x, y = zip(*sorted_pairs)

plt.scatter(x, y)
plt.title("Path on Oscilloscope")
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim([0, DAC_MAX])
plt.ylim([0, DAC_MAX])
plt.grid(True)
plt.show()

#print(x)
print(len(x))
#print(y)
```
